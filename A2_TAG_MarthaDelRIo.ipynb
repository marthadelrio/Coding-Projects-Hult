```python
#using time.sleep() throughout the code to space the prompts and storyline
import time
#using random.choice() throughout the code for lists with multiple elements (to have a different plot when replaying)
import random

#definition of introductory function (welcomes the player to the game and the world it's based on)
def intro():
    print("""
    Welcome to 
  _______ _            __  __                 _____                             
 |__   __| |          |  \/  |               |  __ \                            
    | |  | |__   ___  | \  / | __ _ _______  | |__) _   _ _ __  _ __   ___ _ __ 
    | |  | '_ \ / _ \ | |\/| |/ _` |_  / _ \ |  _  | | | | '_ \| '_ \ / _ | '__|
    | |  | | | |  __/ | |  | | (_| |/ |  __/ | | \ | |_| | | | | | | |  __| |   
    |_|  |_| |_|\___| |_|  |_|\__,_/___\___| |_|  \_\__,_|_| |_|_| |_|\___|_|   
    
    Text Adventure Game
    """)

    input(prompt = """
    Get ready to face the challenges the Maze has to offer. Each prompt
    has a hint on how to answer, the rest is up to you. You got this!
    Press <ENTER> to continue. \n""")

    print("""
    You find yourself waking up in The Glade, surrounded by massive walls 
    that delimit a mysterious maze. You try to remember how you got here, 
    but you can't remember much. """)
    time.sleep(2)

    question1 = input(prompt = """
        Do you remember your name? (<y/n>) """)

    #these are the accepted answers to ask for the name
    answer1_lst = ["Yes", "yes", "Y", "y", "Maybe", "maybe", "Sure", "sure",
                   "Affirmative", "affirmative", "Yes.", "yes.", "Y.", "y.",
                   "Maybe.", "maybe.", "Sure.", "sure.", "Affirmative.", "affirmative."]

    #checking if a match is found
    ans = False

    #for loop to ensure only valid answers are accepted for question1
    for answer1 in answer1_lst :
        if question1.lower() in answer1.lower() :
            #only ask for name if the input is found in the list
            name = input(prompt = """\n
        Please write down your name: \n
        """).title()
            time.sleep(1)

            print(f"""
    Hey, {name}! Welcome to The Glade. Get used to people calling you Greenie 
    from now on. You're our newbie, so suck it up. Your goal is to navigate 
    through the maze and reach the exit, if that's even possible. Be careful, 
    make the right choices, and good luck Greenie! 
    
    \n\n\n""")
            time.sleep(5)
            input(prompt = "Press <ENTER> to continue. \n")
            ans = True
            break
    
    #use if not to not go out of the loop
    if not ans :
        print("""
    Don't worry too much. It might come back later, Greenbean. Get used to 
    people calling you Greenie from now on. You're our newbie, so suck it up.
    Your goal is to navigate through the maze and reach the exit, if that's 
    even possible. Be careful, and make the right choices, and good luck 
    Greenie! 
    
    \n\n\n""")
        time.sleep(5)
        input(prompt = "Press <ENTER> to continue. \n")

#definition of the first stage function
def stage_1() :
  
    print("\n\n Stage 1: THE GLADE \n")
    time.sleep(2)
    
    print("""
    The boys that live in The Glade welcome you. They seem nice enough. They 
    call themselves the Gladers, and they each have a job to complete. It's 
    your turn to choose yours. 
    With a surprising serious tone, a Glader asks you the following riddle: """)
    time.sleep(3)
    
    #list with two riddles for different storylines when replaying
    question2 = ["""
        'Twisted paths and thorny hedges, the entrance and exit are both on 
        the edges. Many choices and dead ends, salvation is found if you 
        stick with your friends. What is it?' (hint: labyrinth synonym)\n""",
        """
        'What is it that one needs most in the long run?' (hint: take a deep,
        deep ______)\n"""]
    
    answer2_lst = ["A maze", "Maze","a maze", "maze", "A maze.", "Maze.", 
                   "a maze.", "maze.", "A labyrinth.", "Labyrinth","a labyrinth", 
                   "labyrinth", "A labyrinth.", "Labyrinth.", 
                   "a maze.", "maze.", "Your breath.", "Your breath", 
                   "your breath", "your breath.", "Breath.", "breath.", 
                   "Breath", "breath"]

    answer2 = input(prompt = random.choice(question2)).lower()

    if answer2 in answer2_lst :
        
        print("""
    You answered correctly, making everyone impressed. You're to head to 
    The Maze, as you have the blood of a runner. Maybe you'll be the first 
    one to finally escape. 
    
    \n\n\n""")
        time.sleep(3)
        input(prompt = "Press <ENTER> to continue. \n")
        return True
    
    else :
        
        print("""
    You gave a wrong answer, which everyone look to be expecting. After a few 
    moments you're new job is announced: \n""")
        time.sleep(3)
        
        jobs_lst = ["builder", "cook", "slicer", "gardener", "med-jack", "map-maker"]
        rand_job_lst = random.choice(jobs_lst)
        
        print(f"""
    Turns out the Gladers where in need of one more {rand_job_lst}. Try to 
    fit in Greenie, and remember the rules of The Glade: \n""")
        time.sleep(3)
        
        print("    1. Everyone does their part. No slackers. \n")
        time.sleep(1)
        
        print("    2. Never hurt another Glader. You have to trust each other. \n")
        time.sleep(1)
        
        print("    3. Never go outside the Glade, you are not a Runner. \n")
        return False
    
#definition of the second stage function
def stage_2():
  
    print("\n\nStage 2: THE MAZE \n")
    time.sleep(2)
    
    print("""
    You enter the cold maze, strange sounds echoing in the distance. 
    As you walk along the corridor, you are presented with two options: go 
    left or right. \n""")
    time.sleep(2)
    
    question3 = input(prompt = "    What are you choosing? (<l/r>) \n")

    if question3 == "left" or question3 == "l" :
        
        print("""
        
    You decide to walk down the left path. After walking for what seems like 
    hours, you hear a big tumbling noise in the distance, as vibrations from 
    the ground reach you. It seems like choosing left was a wise decision. \n""")
        time.sleep(3)
        
        input(prompt = "    Press <ENTER> to continue. \n")

    elif question3 == "right" or question3 == "r" :
        
        print("\n    You decide to walk down the right path. \n")
        time.sleep(2)
        
        print("    It's eerily silent. \n")
        time.sleep(2)
        
        print("    After walking for a couple of minutes, you get a bad feeling. \n")
        time.sleep(2)
        
        print("    Suddenly... \n")
        time.sleep(2)
        
        print("    WHAM! \n")
        time.sleep(1)
        
        #list with three scenarios for different storylines when replaying
        #each scenario consists of and action, an answer, a correct and a wrong answer element
        #multiple elememts are added to ensure randomness when playing
        question4_lst = [
            {"action": """
    The walls start moving! The maze is shifting! Are you: (<a/b/c>): \n 
        a) running back.
        b) running forward.
        c) staying put. \n""", 
             "answers": ["a", "a)", "A", "A)", "a) Running back.", 
                         "a) running back.", "a Running back.", "a running back.", 
                         "Running back.", "Running back", "running back", "b", "b)", 
                         "B", "B)", "b) Running forward.", "b) running forward.", 
                         "b Running forward.", "b running forward.", 
                         "Running forward.", "Running forward", "running forward"], 
             "correct_answer4": ["    Great choice! You avoided getting yourself killed!",
                                 "    Nice! You made it through unscathed.",
                                 "    Impressive! You sure are a Runner."],
            "wrong_answer4": [
            "    Oh no, that was a wrong choice! A wall squashed you to a pulp! Sorry Greenie...",
            "    No! Greenie! As you didn't move a wall ended up collapsing on you..."
           ]},
            {"action": """
        The wall in front of you is collapsing! Watch out! (<Press d to dodge>) """, 
             "answers": ["D", "d", "D.", "d.","Dodge", "dodge", "Dodge.", "dodge."], 
             "correct_answer4": ["    Great! You avoided getting yourself killed!",
                                 "    Nice! You made it through unscathed.",
                                 "    Impressive reflexes! You sure are a Runner."],
            "wrong_answer4": [
            "    Too slow! Sorry Greenie...",
            "    No! Greenie! As you didn't move a wall ended up collapsing on you..."
           ]},
            {"action": """
        You fall flat on your face! You tripped over a crevice... that must hurt... 
        Well, can you stand? (<y/n>)""", "answers": ["Yes", "yes", "Y", "y", "Maybe", 
                                                     "maybe", "Sure", "sure", "Affirmative", 
                                                     "affirmative", "Yes.", "yes.", "Y.", 
                                                     "y.", "Maybe.", "maybe.", "Sure.", 
                                                     "sure.", "Affirmative.", "affirmative."], 
             "correct_answer4": ["    Great! Keep going Greenie!",
                                 "    Nice! You're strong Greenie.",
                                 "    Impressive! You sure are a Runner."],
            "wrong_answer4": [
            "    Oh no, a true Runner can't give up! Sorry Greenie...",
            "    No! Greenie! As you didn't move a wall ended up collapsing on you..."
           ]}
        ]
        
        question4 = random.choice(question4_lst)
        answer4 = input(prompt = question4["action"])

        if answer4.lower() in question4["answers"] :
            
            #display of random encouraging words when answered correectly
            time.sleep(0.5)
            print()
            print(random.choice(question4["correct_answer4"]))
            print()
            time.sleep(2)
            print("""
        
    That was close! Too close! You seem a little shaken Greenie... \n""")
            return True
            
        else :
            
            #display of random failed message when answered incorreectly
            time.sleep(0.5)
            print()
            print(random.choice(question4["wrong_answer4"]))
            print()
            time.sleep(2)
            return False

    #asks user if they want to continue or stop the game
    question5 = "    What will you do? Do you want to keep going? (<y/n>)\n"
    answer5_lst = ["Yes", "yes", "Yes!", "yes!", "Y", "y", "Yeah", "yeah", 
                   "Yeah!", "yeah!", "Sure", "sure", "Sure!", "sure!"]

    answer5 = input(prompt = question5).lower()
    time.sleep(1)

    #if user wants to proceed, moves to the next stage
    if answer5 in answer5_lst :
        
        print("""
    The path ahead is clear, keep moving. 
    
    \n\n\n""")
        time.sleep(2)
        return True
      
    #if user wants to quit, ends game
    else :
        
        print("""
    Although you did your best, the Maze proved to be too much for you, maybe 
    you weren't a Runner after all. \n""")
        time.sleep(2)
        return False

#definition of the third stage function
def stage_3():

    print("\n\nStage 3: THE GRIEVERS\n")
    time.sleep(2)
    
    print("""
    You keep walking through the endless corridors of the Maze, 
    walls on every turn you take. As you approach what seems to be a dead 
    end, you notice a big, protruding, metal door that sticks out of the 
    end wall. \n""")
    time.sleep(3)
    
    print("\n    You found it! A way out! \n")
    time.sleep(2)
    
    print("    In the door, there seems to be some letters engraved: W.C.K.D. \n")
    time.sleep(2)
    
    print("    There's also a lock that needs a code... \n")
    time.sleep(2)
    
    print("    This is too easy. You feel too safe, as if something's not right... \n\n")
    time.sleep(3)
    
    print("    BAM! \n")
    time.sleep(2)
    
    print("""
    Without any notice, you go flying across a corridor from a harsh hit!""")
    time.sleep(2)
    
    print("""
    You've encountered Grievers, lethal and ugly mechanical creatures! You 
    can't fight them, they're too strong! There's only one way out of here, 
    you have to open the door!! \n""")
    time.sleep(3)
  
    question6 = """
    Think! What could the code to unlock the door be? It's four letters! 
    You should know it if you've been paying attention! \n"""
    answer6_lst = ["W.C.K.D.", "WCKD", "w.c.k.d.", "wckd", "W.C.K.D", "w.c.k.d"]
  
    answer6 = input(prompt = question6).lower()
  
    if answer6 in answer6_lst :
        
        print("    Correct! You manage to barely get away from the Grievers.")
        time.sleep(3)
        return True
      
    else :
        
        print("    Incorrect. The Grievers attack is lethal, as expected... Sorry Greenie...")
        time.sleep(3)
        return False

def win() :
    
    print("""\n\n
    Congratulations! 
                    You successfully escaped the Maze! 
                    You were a true Runner at heart! """)
    time.sleep(2)

def fail() :
    
    print("""\n\n
    Game over. 
              The Maze proved to be too challenging for you. 
              Not everyone is born to be a Runner...""")
    time.sleep(2)

#runs the game in just one click
def main() :
    
    #runs the code as long as it doesn't break out of the loop by winning or loosing
    while True :
        
        intro()
        stages = [stage_1, stage_2, stage_3]
        
        # for loop to go through each stage of the game
        for stage in stages :
        
            #fails the game if a stage is not passed
            if not stage() :
                
                fail()
                return
        
        #wins game if all stages are passed
        else :
        
            win()
        
        #asks user if the want to play again
        play_again = input(prompt = "Do you want to play again? (<y/n>): ").lower()
        
        if play_again.lower() not in ["yes", "y", "Yes", "Y", "yes.", "y.", "Yes.", "Y."] :
            
            print("Alrighty Greenie, thanks for playing!")
            break

main()
```
